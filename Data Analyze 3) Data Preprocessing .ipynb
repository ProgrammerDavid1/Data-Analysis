## 데이터 전처리

seaborn에서 제공하는 planets 데이터셋을 바탕으로 실습을 진행해볼 것입니다.

아래 코드와 같이 **`load_dataset`** 함수를 통해 seaborn에서 제공하는 여러 데이터셋을 불러올 수 있습니다.

https://github.com/mwaskom/seaborn-data

해당 링크에서 불러올 수 있는 데이터셋들을 확인할 수 있습니다.

```python
# 데이터셋 준비
import pandas as pd
import seaborn as sns
df = sns.load_dataset('planets')

df.head(5)
```

### 누락 데이터 처리

데이터셋은 여러 이유로 종종 누락된 데이터가 존재할 수 있습니다.

누락 데이터를 처리하지 않으면 분석을 방해하거나 데이터의 품질을 저하하는 등의 문제를 일으킵니다.

이번 실습에서는 이런 결측치를 처리하는 대표적인 두 가지 방법을 소개하겠습니다.

### 1. 누락 데이터가 포함된 행 제거하기

  - Step 1) DataFrame의 **`isnull`** 함수와 **`sum`** 함수를 사용하여 누락된 데이터의 수를 확인하고,
  - Step 2) **`dropna`** 함수를 통해 누락된 열 중 **가장 수가 적은 두 column**들에 대하여 **둘 중 하나라도 누락된** 행들을 제거 해보세요.

  ex) 열 별 누락된 개수가 `number: 100, mass: 6, year: 30`이라면
  mass와 year 열에 대해, 둘 중 하나라도 누락된 데이터를 찾아 제거

**Step 1 Hint)**
- **`isnull`** 함수는, 각 데이터가 NaN인지 확인해서 True 또는 False로 나타내줍니다.
- **`sum`** 함수는, axis 매개변수를 통해 열 또는 행별로 합계를 구해줍니다.
- 수로 나타내면 True는 1이고, False는 0이 되는 것을 통해 데이터의 총 NaN 개수를 알 수 있습니다.

```python
# Step 1. isnull 함수와 sum 함수를 통해 누락 데이터 확인
print("열별 누락된 데이터 개수")
print("=======================")
print(df.isnull().sum(axis=0))
print("=======================")
```

**Step 2 Hint)**
- dropna 함수는 누락된 데이터를 제거할 때 사용되는 함수입니다.
- 매개변수 subset은 어떤 열을 기준으로 결측치를 확인할 지 정해줍니다.\
  여러 열을 기준으로 하려면 리스트로 넣을 수 있습니다.
- 매개변수 how가 "any"면, 기준 중 아무 하나를 만족하면 제거합니다.\
  '"all"이면, 기준 모두를 만족해야 제거합니다.

```python
# Step 2. dropna 함수를 통해 특정 열에 대해서만 누락 데이터 제거
dropped_df = df.dropna(subset = ['orbital_period', 'distance'], how = 'any', axis = 0)

print(dropped_df)
# 출력의 마지막이 [776 rows x 6 columns] 가 나와야 정답입니다.
```

이런 방법 외에도 DataFrame의 **`info`** 함수를 통해, 각 열에 Non-Null 데이터가 몇 개인지 알 수 있습니다.

이를 바탕으로 현재 **mass** 열에도 아직 Null 데이터가 존재함을 알 수 있습니다.

```python
dropped_df.info()
```

```python
dropped_df.tail(5)
```

### 2. 누락 데이터를 대체 값으로 채우기

대체 값으로 채우는 방법은 다양합니다.

통계값을 사용하거나, 이전 값/다음 값으로 채우기도 하고, 알고리즘이나 인공지능을 이용해 채우기도 합니다.

</br>

이번에는 mass 열의 NaN 데이터를 **평균값으로 대체**해볼 것입니다.

- Step 1) 먼저, mass 열의 평균값을 계산하고,
- Step 2) **`fillna`** 함수를 통해 NaN 데이터들을 평균값으로 대체해 보세요.

**주의)** **filled_df**를 바탕으로 진행하세요

**Q)** 결측치가 있는 열에 대해 mean을 해버리면 평균이 잘못 계산되지 않을까요?

**A)** mean 같은 통계 함수에는 자동으로 skipna 라는 매개변수가 True로 들어가게 됩니다.\
이는 결측치를 제외하고 계산하겠다는 의미입니다. 그렇기에 제대로 데이터에 대해서만 평균이 구해지게 됩니다.

```python
filled_df = dropped_df.copy()

# Step 1. 통계함수를 통해 mass 열의 평균값 계산
imputation = filled_df['mass'].mean(axis=0)

print(f"mass열의 평균치: {imputation}")
```

```python
# Step 2. fillna 함수를 통해 distance 열의 NaN 데이터들을 평균값으로 대체
filled_df['mass'] = filled_df['mass'].fillna(imputation)

# inplace = True를 imputation 뒤에 넣을거면 변수로 따라 만들지 않고 하면됨. inplace는 새 데이터가 아닌 원본 자체를 바꾸는거임

# 누락 데이터 채운 후
filled_df.tail(5)
```

### 선택 사항)

- 다양한 통계함수를 사용해보세요.\
  통계함수 중에는 **평균값(mean), 중앙값(median), 최빈값(mode)**이 주로 사용됩니다.

## 정규화
변수가 갖는 범위의 크기는 변수마다 다를 수 있습니다.

이런 경우에는 데이터 분석할 때, 큰 범위를 갖는 변수가 더 큰 영향을 갖게 됩니다.

그래서 보통 모든 변수들이 같은 크기의 범위를 갖도록 **정규화**하는 작업을 가집니다.


```python
df = sns.load_dataset('planets')

df.dropna(subset=['mass', 'distance'], axis=0, inplace=True)

df.describe() # 데이터의 정보를 요약해서 보여주는 함수
```

위의 결과에서 세 가지의 연속형 변수를 살펴봅시다.

oribital_period는 범위가 `1.32` ~ `17337.5`인 것에 비해,\
mass는 `0` ~ `25`, distance는 `1.35` ~ `354`로 범위의 크기가 각자 다르다는 것을 알 수 있습니다.

데이터 분석을 위해선, 이 변수들의 범위를 모두 일정하게 맞춰야할 필요가 있습니다.

</br>


해당 실습에서는 distance 열에 대해서만 정규화를 진행해보도록 하겠습니다.

정규화 방법으로는 데이터의 범위를 0~1으로 변환하는 Min-Max Normalization을 사용해봅시다.

```python
#Step 1. 최솟값으로 빼서 가장 작은 값이 0이 되도록 변환
df['distance'] -= df['distance'].min()
# 예를 들어 3~ 10까지면 최솟값인 3을 빼주면 0~ 7까지로 되겠지

#Step 2. 데이터의 범위 크기를 측정
min_max = df['distance'].max() - df['distance'].min()
# 0~7이면 7 - 0을 해서 7이 되는거지

#Step 3. distance 열을 방금 구한 크기로 나눠 줌
df['distance'] = df['distance'] / min_max
# 0~ 7이니까 7로 나눠주면 0~1까지로 되겠지

print(df['distance'].describe())
# min과 max가 0과 1이어야 정답입니다.
```
