### 필터링
4주차 실습 때 나왔던 numpy의 Boolean Indexing과 유사한 개념입니다.

데이터프레임에 조건식을 이용하여 원하는 데이터만 필터링 할 수 있습니다.

````python
orbit_mask = df['orbital_period'] < 500
print(f"필터: \n{orbit_mask}")

print("\n필터링 된 데이터프레임: ")
df[orbit_mask].head(5)
#df[df['orbital_period'] < 500].head(5)
````

여러 개의 조건을 모두 만족하는 데이터를 필터링 하려면\
조건을 & 연산자로 묶어야 합니다.

해당 실습에서 distance는 0.3보다 크고, mass는 1보다 작은 데이터를 필터링 해보세요.

````python
mask_distance = df['distance'] > 0.3
mask_mass = df['mass'] < 1

df[mask_distance & mask_mass]
````

여러 조건을 하나라도 만족하는 데이터를 필터링 하려면\
조건을 | 연산자로 묶어야 합니다.

해당 실습에서 year이 1995, 1996, 1997 중 하나인 데이터를 필터링 해보세요.

````python
mask_1995 = df['year'] == 1995
mask_1996 = df['year'] == 1996
mask_1997 = df['year'] == 1997

df[mask_1995 | mask_1996 | mask_1997]
````

위와 같이 특정 변수가 여러가지 값 중 하나인 경우인지를 알고 싶다면\
isin 메소드를 사용할 수도 있습니다.

isin 메소드를 이용하여 방금과 같은 조건의 데이터를 필터링 해보세요.

````python
filter_year = df['year'].isin([1995, 1996, 1997])

# mask 변수를 이렇게 3개로 안 나눠도 괜찮구 , 인덱싱으로 넣어줘야함

df[filter_year]
````

### 함수 매핑
데이터에 행 단위, 열 단위, 요소 단위로 특정 연산을 적용하고 싶을 수도 있습니다.

이런 경우에는 apply, applymap 메소드를 이용할 수 있습니다.

 **※ 되도록 따로 함수를 만들지 말고 lambda 함수를 이용하여 해결해봅시다. (람다 함수의 정의를 자세히 읽어 볼 것)**

````python
import pandas as pd

sample_data = {'a':[1,2,3], 'b':[6,7,8]}

print("원본 데이터프레임")
print(pd.DataFrame(sample_data))
````

````python
df = pd.DataFrame(sample_data)
print("원본 데이터프레임")
print(df)

print("================")

# 1. 각 열을 출력하기
df.apply(lambda s: print(s), axis=0)
# apply 메소드는

print("================")

# 2. 각 열의 총합을 출력하기
def sum_column(column):
  return print(column.iloc[0] + column.iloc[1] + column.iloc[2])
df.apply(lambda s: sum_column(s), axis=0)

print("================")

# 3. 각 데이터를 출력하기
df.apply(lambda s: print(s), axis=1)

print("================")

# 4.
def sum_data(data):
  return print(data['a'] + data['b'])
df.apply(lambda s: sum_data(s), axis=1)

print()
````

### 데이터프레임 연결

따로 떨어져 있는 데이터프레임들을 하나로 연결해야 하는 상황이 올 수도 있습니다.

이런 경우는 concat 메소드를 이용할 수 있습니다.

````python
# 데이터셋 준비
import pandas as pd
df1 = pd.DataFrame({'a': ['o', 'o', 'o', 'o'],
                    'b': ['o', 'o', 'o', 'o'],
                    'c': ['o', 'o', 'o', 'o']},
                    index=[0, 1, 2, 3])
df2 = pd.DataFrame({'a': ['-', '-'],
                    'b': ['-', '-'],
                    'c': ['-', '-']},
                    index=[0, 1])
df3 = pd.DataFrame({'d': ['l', 'l', 'l', 'l'],
                    'e': ['l', 'l', 'l', 'l']},
                    index=[0, 1, 2, 3])

# 데이터셋 출력
print('기본 데이터', df1, '', sep='\n')
print('추가 행', df2, '', sep='\n')
print('추가 열', df3, '', sep='\n')
````

위 데이터셋들을 concat 메소드로 아래 예시와 같은 결과가 나오도록 연결해보세요.

- 정답 예시)
```
     a  b  c
0  o  o  o
1  o  o  o
2  o  o  o
3  o  o  o
0  -  -  -
1  -  -  -

     a  b  c
0  o  o  o
1  o  o  o
2  o  o  o
3  o  o  o
4  -  -  -
5  -  -  -

     a  b  c  d  e
0  o  o  o  l  l
1  o  o  o  l  l
2  o  o  o  l  l
3  o  o  o  l  l
```




**Hint)**
- axis 매개변수에 따라서 연결하는 방향을 정할 수 있습니다.

- 연결한 후의 인덱스가 초기화되기 원한다면 ignore_index 매개변수를 활용할 수 있습니다.

````python
# Prob 1. 기본 데이터 df1에 추가 행 df2을 연결해보세요.
result1 = pd.concat([df1,df2], axis=0)
print(result1, '\n')

# Prob 2. result1의 index가 0,1,2,3,0,1 처럼 기존 index를 따라가지 않고, 별개의 순차적인 index를 갖도록 하세요.
# index가 마음에 안 들면 ignore_index = True로 하면 된다.
result2 = pd.concat([df1,df2],ignore_index=True)
print(result2, '\n')

# Prob 3. 기본 데이터 df1에 추가 열 df3을 연결해보세요.
result3 = pd.concat([df1,df3],axis=1)
print(result3)
````

### 데이터프레임 병합

데이터프레임 병합은 연결과는 조금 다른 개념입니다.

두 데이터프레임을 하나로 합친다는 동작은 같지만, 특정 기준에 의해서 합쳐지는 것이 차이점입니다.

기본적으로 그 기준은 양쪽 데이터프레임에 동시에 존재하는 열이 됩니다.

merge 메소드를 이용해 가격 정보를 갖는 price와 재고 정보를 갖는 stock을 병합해봅시다.

````python
# 데이터셋 준비
import pandas as pd
price = pd.DataFrame({'name': ['콜라', '생수', '우유', '녹차'],
                    'price': [1200, 1000, 1300, 900]})

stock = pd.DataFrame({'name': ['우유', '생수', '김밥'],
                    'stock': [2, 10, 7]})

# 데이터셋 출력
print('가격', price, '', sep='\n')
print('재고', stock, '', sep='\n')
````

위 데이터셋들을 merge 메소드로 아래 예시와 같은 결과가 나오도록 연결해보세요.

- 정답 예시)
```
      name  price  stock
0   생수   1000     10
1   우유   1300      2

      name   price  stock
0   콜라  1200.0    NaN
1   생수  1000.0   10.0
2   우유  1300.0    2.0
3   녹차   900.0    NaN
4   김밥     NaN    7.0

      name   price  stock
0   우유  1300.0      2
1   생수  1000.0     10
2   김밥     NaN      7
```

````python
# Prob 1. 두 데이터 price와 stock을 기본 옵션으로 병합하고 결과를 확인하세요.
res1 = pd.merge(price,stock)
print(res1, '\n')

# Prob 2. 위 방법은 name이 양쪽에 공통으로 없는 데이터는 제거해버립니다. how 매개변수를 이용하여 제거하지 않도록 해보세요.
res2 = pd.merge(price,stock, how = 'outer')
print(res2, '\n')

# Prob 3. 재고(stock) 데이터가 있는 물품을 기준으로 병합하고 싶은 경우는 어떻게 해야하는가?
res3 = pd.merge(price, stock, how = 'right')  # right는 오른쪽에 있는 stock을 의미
print(res3, '\n')
````
